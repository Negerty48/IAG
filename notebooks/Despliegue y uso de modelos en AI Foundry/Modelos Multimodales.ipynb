```python
import base64
from openai import AzureOpenAI

client = AzureOpenAI(
    api_key="FzjGvONPceyRYAMQcbB8SfkIIeRvY60Cka1d8yY2KPTf48WJr7tuJQQJ99CDAC5RqLJXJ3w3AAAAACOG0rYa",
    azure_endpoint="https://01-iag.services.ai.azure.com/",
    api_version="2024-08-01-preview"
)

def analizar_imagen(ruta_imagen):    
    with open(ruta_imagen, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": "¿Qué importancia histórica tiene esta foto?"},
                    {
                        "type": "image_url",
                        "image_url": {"url": f"data:image/jpeg;base64,{encoded_string}"}
                    }
                ]
            }
        ]
    )
    return response.choices[0].message.content

print(analizar_imagen("foto_ruinas.jpg"))
```

## Audio

```python
import base64
from openai import AzureOpenAI

client = AzureOpenAI(
    api_key="FzjGvONPceyRYAMQcbB8SfkIIeRvY60Cka1d8yY2KPTf48WJr7tuJQQJ99CDAC5RqLJXJ3w3AAAAACOG0rYa",
    azure_endpoint="https://01-iag.services.ai.azure.com/",
    api_version="2024-08-01-preview"
)

def analizar_audio(ruta_audio):    
    with open(ruta_audio, "rb") as audio_file:
        encoded_audio = base64.b64encode(audio_file.read()).decode('utf-8')

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "user",
                "content": [
                    {"type": "text", "text": "Explica qué importancia histórica tiene este audio y resume qué se dice."},
                    {
                        "type": "input_audio",
                        "input_audio": {
                            "data": encoded_audio,
                            "format": "mp3" 
                        }
                    }
                ]
            }
        ]
    )
    return response.choices[0].message.content

print(analizar_audio("AUD-20260227-WA0012.mp3"))
```
